```python
import numpy as npnp.random.seed(42)x = np.random.random(10)y = np.sin(2 * x)p = np.polyfit(x, y, 1)  # fit a 1st-degree polynomial (i.e. a line) to the dataprint pxfit = np.linspace(0, 1, 10)yfit = np.polyval(p, xfit)import pylab as plpl.scatter(x, y,color="black")pl.plot(xfit, yfit)pl.xlabel('x')pl.ylabel('y')pl.title('Linear Regression Example')#Test our fitx_new = np.random.random(50)y_new = np.polyval(p, x_new)  # evaluate the polynomial at x_newprint "Average Error is %s" % np.mean(abs(np.sin(2 * x_new) - y_new))
```
